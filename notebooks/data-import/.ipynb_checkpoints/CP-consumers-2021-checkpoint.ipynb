##### Combine DataFrames

```python
prices_df_list = [
    prices_jan_2021, prices_feb_2021, prices_mar_2021, prices_apr_2021,
    prices_may_2021, prices_jun_2021, prices_jul_2021, prices_aug_2021,
    prices_sep_2021, prices_oct_2021, prices_nov_2021, prices_dec_2021
]
```

```python
prices_2021 = pd.DataFrame(columns=prices_jan_2021.columns)
for df in prices_df_list:
    prices_2021 = pd.merge(prices_2021.T, df.T, on=prices_2021.columns)
    prices_2021.index = prices_2021['key_0']
    prices_2021.index.name = None
    prices_2021 = prices_2021.drop('key_0', axis=1)
    prices_2021 = prices_2021.T
prices_2021.columns = [i.lower() for i in prices_2021.columns]
```
